```python
def outliers(column,y):
    l=[]
    q1=y[column].quantile(0.25)
    q3=y[column].quantile(0.75)
    q2=y[column].quantile(0.5)
    iqr=(q3-q1)*1.5
    upper=iqr+q3
    lower=q1-iqr
    for i in y[column]:
        if i <upper and i >lower:
            l.append(i)
        else:
            l.append(q2)
    return l
```

```python
wi_drop=wi.drop(columns="Proline")
for i in wi_drop:
    wi_drop[i]=outliers(i,wi_drop)
```

```python
sns.boxplot(data=wi_drop)
plt.xticks(rotation=270)
```
